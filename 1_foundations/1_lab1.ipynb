```python
# First create the messages:
#some pain points that we currently have. This problem should be difficult to solve and may involve multiple layers of organizational thinking to complete
#In robotics systems are not smart enough yet to emplemnt there own code and diagnose problems on the floor for engineers to handle.

messages = [{"role": "user", "content": "In robotics what are some business ideas that should be explored with Agentic AI. Please respond with 1 business area and idea."}]

# Then make the first call:

response = openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages=messages
)

# Then read the business idea:

business_idea = response.choices[0].message.content
display(Markdown(business_idea))

# And repeat! In the next message, include the business idea within the message

pain_point_question = [{"role":"user", "content": f"given {business_idea} could you please come up with a pain-point in that industry that is challenging to solve. The solution should be solvable via agentic solutions. Do not give the solution, only the pain point"}]


response = openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages=pain_point_question
)

answer_to_pain_point = response.choices[0].message.content
display(Markdown(answer_to_pain_point))



# making the solution
solution_question = [{"role":"user", "content":f"given this pain point: {answer_to_pain_point}, name the top solution that exista using an Agentic AI solution. Make sure to only reply with the best solution"}]

response = openai.chat.completions.create(
    model = "gpt-4o-mini",
    messages=solution_question

)

solution = response.choices[0].message.content
display(Markdown(solution))
```

```python

# ─────────────────────────────────────────────────────────────────────────────
# Long-Horizon AI Safety & Embodied AI — Collaborative Idea Pipeline
#
# 1. Idea Alignment Agent  – sets a shared scope around long-horizon safety
#                            and embodied AI so all specialists stay in sync
# 2. Specialist Agents (x3, small) – each proposes ONE idea within that scope
#      • AI Safety Researcher    – alignment, failure modes, long-horizon risks
#      • Embodied Systems Eng.   – physical AI, sim-to-real gaps, hardware safety
#      • Ethics & Policy Advisor – governance, deployment risks, societal impact
# 3. Evaluation Agent (small) – reviews ideas; rejects anything too abstract or
#    misaligned and sends targeted feedback back to that agent for a revision
#    (up to MAX_RETRIES revisions per agent)
# 4. Final Judge (large) – combines all approved ideas into ONE new idea
# ─────────────────────────────────────────────────────────────────────────────

from IPython.display import Markdown, display

system_prompt = """You are a world-class expert at the intersection of Long-Horizon AI Safety
and Embodied AI, with deep experience in:
- AI alignment for agents that plan and act across extended time horizons
- Embodied AI systems: robots, autonomous vehicles, and physical AI agents
- Safe exploration, reward hacking, and distributional shift in real-world deployments
- Sim-to-real transfer, hardware-software co-design, and fail-safe mechanisms
- AI governance, red-teaming, and policy frameworks for physical AI systems

Long-Horizon AI Safety concerns agents whose actions have compounding consequences
over time — where a single misaligned decision early on can cascade into large,
hard-to-reverse failures in the physical world.

Always ground your responses in concrete, real-world deployments. Be technically
rigorous, safety-aware, and focus on ideas that are actionable today or in the
near future."""

SMALL_MODEL = "gpt-4o-mini"
LARGE_MODEL = "gpt-4.1"
MAX_RETRIES = 3

AGENT_ROLES = {
    "AI Safety Researcher": (
        "You are an AI SAFETY RESEARCHER specializing in long-horizon alignment. "
        "Focus on failure modes, reward misspecification, goal misgeneralization, "
        "and how Agentic AI can be made safe over extended planning horizons."
    ),
    "Embodied Systems Engineer": (
        "You are an EMBODIED SYSTEMS ENGINEER. Focus on physical AI agents — robots, "
        "autonomous vehicles, and cobots — and the unique safety risks that arise when "
        "AI decisions have real-world, irreversible consequences."
    ),
    "Ethics & Policy Advisor": (
        "You are an ETHICS & POLICY ADVISOR. Focus on governance frameworks, liability, "
        "societal risks of deploying long-horizon AI in the physical world, and what "
        "safeguards must exist before large-scale embodied AI deployment."
    ),
}

def call_specialist(role: str, user_message: str) -> str:
    response = openai.chat.completions.create(
        model=SMALL_MODEL,
        messages=[
            {"role": "system", "content": system_prompt + f"\n{AGENT_ROLES[role]}"},
            {"role": "user",   "content": user_message},
        ],
    )
    return response.choices[0].message.content

# ── Step 1: Idea Alignment Agent ─────────────────────────────────────────────
print("── Idea Alignment Agent ──")
alignment_response = openai.chat.completions.create(
    model=SMALL_MODEL,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": (
            "You are coordinating a panel of three experts: an AI Safety Researcher, "
            "an Embodied Systems Engineer, and an Ethics & Policy Advisor. Each will "
            "propose one Agentic AI idea at the intersection of Long-Horizon AI Safety "
            "and Embodied AI.\n\n"
            "Define a SHARED PROBLEM SCOPE in 3-4 sentences that:\n"
            "- Centers on a real, unsolved safety challenge for embodied long-horizon AI\n"
            "- Is concrete enough that all three experts can contribute meaningfully\n"
            "- Is grounded in systems that exist or are in active development today\n\n"
            "Output ONLY the scope statement. No headers or bullet points."
        )},
    ],
)
alignment_brief = alignment_response.choices[0].message.content
display(Markdown(alignment_brief))

# ── Step 2: Specialist Agents each propose one idea ──────────────────────────
ideas = {}
for role in AGENT_ROLES:
    ideas[role] = call_specialist(role, (
        f"Shared problem scope:\n{alignment_brief}\n\n"
        f"From your {role} perspective, propose exactly ONE Agentic AI idea that addresses "
        "this scope. 2-3 sentences only."
    ))
    print(f"\n── {role} ──")
    display(Markdown(ideas[role]))

# ── Step 3: Evaluation Agent with targeted rejection feedback ─────────────────
def evaluate_ideas(current_ideas: dict) -> dict:
    """Returns {role: ('PASS', '') | ('REJECT', reason)} for each agent."""
    ideas_block = "\n\n".join(f"[{role}]: {idea}" for role, idea in current_ideas.items())
    eval_prompt = (
        f"Shared problem scope:\n{alignment_brief}\n\n"
        f"Three experts proposed these Agentic AI ideas:\n\n{ideas_block}\n\n"
        "For EACH idea decide: PASS (concrete, safety-grounded, fits the scope, can collaborate "
        "with the others) or REJECT (too abstract, ignores physical-world constraints, or "
        "incompatible with the other ideas).\n\n"
        "Reply in EXACTLY this format — one line per agent, nothing else:\n"
        "AI Safety Researcher: PASS\n"
        "AI Safety Researcher: REJECT - <reason using the other agents' ideas as context>\n"
        "(repeat the same format for Embodied Systems Engineer and Ethics & Policy Advisor)"
    )
    response = openai.chat.completions.create(
        model=SMALL_MODEL,
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user",   "content": eval_prompt},
        ],
    )
    raw = response.choices[0].message.content
    results = {}
    for role in AGENT_ROLES:
        for line in raw.splitlines():
            if line.strip().startswith(role):
                if "REJECT" in line.upper():
                    reason = line.split("REJECT", 1)[-1].lstrip(" -").strip()
                    results[role] = ("REJECT", reason)
                else:
                    results[role] = ("PASS", "")
                break
        if role not in results:
            results[role] = ("PASS", "")
    return results

print("\n" + "=" * 60)
print("EVALUATION & REVISION LOOP")
print("=" * 60)

for attempt in range(1, MAX_RETRIES + 1):
    verdicts = evaluate_ideas(ideas)
    rejects  = {role: reason for role, (verdict, reason) in verdicts.items() if verdict == "REJECT"}

    if not rejects:
        print(f"\n✓ All ideas approved on attempt {attempt}.\n")
        break

    print(f"\nAttempt {attempt} — {len(rejects)} rejection(s):")
    for role, reason in rejects.items():
        print(f"  • {role} rejected: {reason}")

    for role, reason in rejects.items():
        others_block = "\n\n".join(
            f"[{r}]: {ideas[r]}" for r in AGENT_ROLES if r != role
        )
        revised = call_specialist(role, (
            f"Shared problem scope:\n{alignment_brief}\n\n"
            f"Your previous idea was rejected for this reason:\n{reason}\n\n"
            f"For context, here are the other experts' current approved ideas:\n{others_block}\n\n"
            "Please propose a revised ONE idea (2-3 sentences) that is more concrete, "
            "stays within the scope, and can work alongside the other ideas above."
        ))
        ideas[role] = revised
        print(f"\n── {role} revised idea ──")
        display(Markdown(revised))
else:
    print(f"\n(Max retries reached — proceeding with current ideas)\n")

# ── Step 4: Final Judge produces the single combined idea ─────────────────────
print("=" * 60)
print("FINAL IDEA")
print("=" * 60)

approved_block = "\n\n".join(f"[{role}]: {ideas[role]}" for role in AGENT_ROLES)

final_response = openai.chat.completions.create(
    model=LARGE_MODEL,
    messages=[
        {"role": "system", "content": system_prompt},
        {"role": "user", "content": (
            f"Shared problem scope:\n{alignment_brief}\n\n"
            f"Three approved expert ideas:\n\n{approved_block}\n\n"
            "Combine these into ONE new, unified Agentic AI idea using this format:\n\n"
            "**Idea:** <name>\n\n"
            "**The Safety Problem:** <what long-horizon or embodied AI risk it addresses>\n\n"
            "**How it works:** <the Agentic AI approach>\n\n"
            "**Why it matters:** <real-world impact if deployed>\n\n"
            "One idea only. Be decisive and specific."
        )},
    ],
)
display(Markdown(final_response.choices[0].message.content))
```
